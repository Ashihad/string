```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import math
%matplotlib qt
```

```python
N_x = 101
dx = 0.01
dt = 0.005
xmax = 1
tmax = 5
N_t = int(tmax/dt)
```

```python
# verlet integration method

x = np.empty(N_x)
for i in range(N_x):
    x[i] = i*dx
t_arr = np.empty(N_t)
for i in range(N_t):
    t_arr[i] = i*dt
# format is (time, x)
u = np.empty((N_t, N_x))
v = np.empty((N_t, N_x))
a = np.empty((N_t, N_x))
# initial u and v at t = 0
for i in range(N_x):
    u[0][i] = np.exp(-100*(i*dx-0.5)**2)
    v[0][i] = 0
for i in range(N_x):
    if i == 0 or i == N_x-1:
        a[0][i] = 0
    else:
        a[0][i] = (u[0][i-1] + u[0][i+1] -2*u[0][i])/dx**2
# iteration over time
for t in range(1, N_t):
    # iteration over string, get u
    for i in range(N_x):
        # bounds
        if i == 0 or i == N_x-1:
            u[t][i] = 0
        else:
            u[t][i] = u[t-1][i] + dt*v[t-1][i] + dt**2/2*a[t-1][i]
    # a
    for i in range(N_x):
        if i == 0 or i == N_x-1:
            a[t][i] = 0
        else:
            a[t][i] = (u[t][i+1] + u[t][i-1] - 2*u[t][i])/dx**2
    # v
    for i in range(N_x):
        if i == 0 or i == N_x-1:
            v[t][i] = 0
        else:
            v[t][i] = v[t-1][i] + dt/2*(a[t][i] + a[t-1][i])

fig = plt.figure()
plt.imshow(u.T, cmap='afmhot', interpolation = 'nearest', aspect='auto')
plt.colorbar()
plt.xlabel('czas')
plt.ylabel('pozycja')
plt.show()
```
